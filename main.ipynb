```python
import pandas as pd
from kloppy import impect
```

```python
match_id = 122838 
dataset = impect.load_open_data(
    match_id=match_id,
    competition_id=743,
)
```

```python
dataset.transform(to_coordinate_system="secondspectrum")  
dataset.filter(lambda event: event.event_type.name in ["PASS", "SHOT"])
dataset.to_df(engine="polars")  # or engine="pandas"
```
